# Rename DICOM Folders Based on Demographics

This notebook renames subfolders in the DICOM directory based on the Name-to-ID mapping from the ParticipantsDemographics.xlsx file.

**Important:** Make sure to backup your data before running this script!

```python
import os
import pandas as pd

# Paths
dicom_root_path = r"E:/Muscle/data_sherpas/DICOMsAnon"
demographics_file = r"D:/data_sherpas/ParticipantsDemographics.xlsx"

# Load demographics data
print("Loading demographics file...")
demographics_df = pd.read_excel(demographics_file, sheet_name='Sheet1')

# Display the columns to verify
print(f"\nColumns in demographics file: {demographics_df.columns.tolist()}")
print(f"\nFirst few rows:")
print(demographics_df[['Filenames', 'ID']].head(10))
```

```python
# Create a mapping dictionary: Name -> ID
name_to_id = dict(zip(demographics_df['Filenames'], demographics_df['ID']))

print(f"\nCreated {len(name_to_id)} name-to-ID mappings")
print("\nMapping (Name -> ID):")
print("="*60)
print(name_to_id.items())
for name, id_val in sorted((k, v) for k, v in name_to_id.items() if pd.notna(k)):
    print(f"  '{name}' -> '{id_val}'")
```

```python
# List all subfolders in the DICOM directory
print(f"\nChecking directory: {dicom_root_path}")

if os.path.exists(dicom_root_path):
    subfolders = [f for f in os.listdir(dicom_root_path) if os.path.isdir(os.path.join(dicom_root_path, f))]
    print(f"Found {len(subfolders)} subfolders\n")
    
    print("Current folder names:")
    print("="*60)
    for folder_name in sorted(subfolders):
        # Check if it matches any name in the mapping
        if folder_name in name_to_id:
            print(f"  ✓ '{folder_name}' -> will rename to '{name_to_id[folder_name]}'")
        else:
            print(f"  ✗ '{folder_name}' -> NOT FOUND in demographics")
else:
    print(f"ERROR: Directory not found: {dicom_root_path}")
```

## Perform the Renaming

**⚠️ WARNING:** The cell below will rename the folders. Make sure you have reviewed the mapping above and have a backup!

Run the cell below only when you're ready to perform the renaming operation.

```python
# Perform the renaming operation
if os.path.exists(dicom_root_path):
    subfolders = [f for f in os.listdir(dicom_root_path) if os.path.isdir(os.path.join(dicom_root_path, f))]
    
    # Track renaming operations
    renamed_count = 0
    not_found_count = 0
    already_named_count = 0
    errors = []
    
    print("Starting renaming operations...")
    print("="*60)
    
    for folder_name in subfolders:
        old_path = os.path.join(dicom_root_path, folder_name)
        
        # Check if folder name matches any Name in the demographics
        if folder_name in name_to_id:
            new_folder_name = name_to_id[folder_name]
            new_path = os.path.join(dicom_root_path, new_folder_name)
            
            # Check if the new name already exists
            if os.path.exists(new_path):
                print(f"  SKIP: '{folder_name}' -> '{new_folder_name}' (target already exists)")
                already_named_count += 1
            else:
                try:
                    # Rename the folder
                    os.rename(old_path, new_path)
                    print(f"  ✓ RENAMED: '{folder_name}' -> '{new_folder_name}'")
                    renamed_count += 1
                except Exception as e:
                    error_msg = f"Error renaming '{folder_name}': {str(e)}"
                    print(f"  ✗ ERROR: {error_msg}")
                    errors.append(error_msg)
        else:
            print(f"  ⚠ NOT FOUND: '{folder_name}' (no matching Name in demographics)")
            not_found_count += 1
    
    print(f"\n{'='*60}")
    print(f"SUMMARY:")
    print(f"{'='*60}")
    print(f"  Total folders processed: {len(subfolders)}")
    print(f"  Successfully renamed: {renamed_count}")
    print(f"  Already correctly named (skipped): {already_named_count}")
    print(f"  Not found in demographics: {not_found_count}")
    print(f"  Errors: {len(errors)}")
    
    if errors:
        print(f"\nErrors encountered:")
        for error in errors:
            print(f"  - {error}")
    
    print(f"{'='*60}")
else:
    print(f"ERROR: Directory not found: {dicom_root_path}")
```

## Anonymize DICOM Files

**⚠️ WARNING:** The cell below will modify DICOM files by replacing patient names with the folder ID.

This will:
- Go through each folder in the DICOM directory
- Find all DICOM files (recursively in subfolders)
- Replace patient name tags with the folder ID (anonymization)

Make sure you have a backup before running!

```python
import pydicom
import glob
from pathlib import Path

# Anonymize DICOM files in each folder
if os.path.exists(dicom_root_path):
    # Get all folders (should now have ID names after renaming)
    folders = [f for f in os.listdir(dicom_root_path) if os.path.isdir(os.path.join(dicom_root_path, f))]
    
    print(f"Found {len(folders)} folders to process for anonymization")
    print("="*60)
    
    total_files_processed = 0
    total_files_anonymized = 0
    total_errors = 0
    
    for folder_name in sorted(folders):
        folder_path = os.path.join(dicom_root_path, folder_name)
        
        # Find all DICOM files recursively in this folder
        # DICOM files typically have .dcm extension or no extension
        dicom_files = []
        
        # Search for files in the folder and subfolders
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                # Try to identify DICOM files (they may have .dcm extension or no extension)
                if file.lower().endswith('.dcm') or '.' not in file:
                    dicom_files.append(file_path)
        
        if len(dicom_files) == 0:
            print(f"  ⚠ {folder_name}: No DICOM files found")
            continue
        
        print(f"\n  Processing folder: {folder_name}")
        print(f"    Found {len(dicom_files)} potential DICOM files")
        
        files_anonymized = 0
        files_skipped = 0
        errors = 0
        
        for dicom_file_path in dicom_files:
            try:
                # Read the DICOM file
                ds = pydicom.dcmread(dicom_file_path, force=True)
                
                # Check if it's a valid DICOM file
                if not hasattr(ds, 'PatientName'):
                    files_skipped += 1
                    continue
                
                # Anonymize: Replace patient identifiable information with folder ID
                ds.PatientName = folder_name
                ds.PatientID = folder_name
                
                # Optionally anonymize other fields
                if hasattr(ds, 'PatientBirthDate'):
                    ds.PatientBirthDate = ''
                if hasattr(ds, 'PatientSex'):
                    pass  # Keep sex for analysis
                if hasattr(ds, 'PatientAge'):
                    pass  # Keep age for analysis
                if hasattr(ds, 'OtherPatientNames'):
                    ds.OtherPatientNames = ''
                
                # Save the anonymized DICOM file
                ds.save_as(dicom_file_path)
                files_anonymized += 1
                
            except Exception as e:
                errors += 1
                if errors <= 3:  # Only print first 3 errors per folder
                    print(f"      ✗ Error processing {os.path.basename(dicom_file_path)}: {str(e)}")
        
        print(f"    ✓ Anonymized: {files_anonymized}, Skipped: {files_skipped}, Errors: {errors}")
        
        total_files_processed += len(dicom_files)
        total_files_anonymized += files_anonymized
        total_errors += errors
    
    print(f"\n{'='*60}")
    print(f"ANONYMIZATION SUMMARY:")
    print(f"{'='*60}")
    print(f"  Total folders processed: {len(folders)}")
    print(f"  Total files found: {total_files_processed}")
    print(f"  Total files anonymized: {total_files_anonymized}")
    print(f"  Total errors: {total_errors}")
    print(f"{'='*60}")
else:
    print(f"ERROR: Directory not found: {dicom_root_path}")
```
